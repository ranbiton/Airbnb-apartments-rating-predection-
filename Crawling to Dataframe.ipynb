### function that count and save all the url's of the apartments in one page

```python

def CountPageUrls(driver):
    currentPageApartmentsUrl = []
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    prefix_url="https://www.airbnb.com"
    for row in soup.findAll("div", attrs={'class':"cy5jw6o dir dir-ltr"}):
        currentPageApartmentsUrl.append(prefix_url+row.find("a")['href'])
    return currentPageApartmentsUrl
print(CountPageUrls(driver))
```

### Lists initialization

```python
apartments_names = []
prices_per_month= []
overall_rating = []
num_of_reviews = []
num_of_guests=[]
accuracy_score=[]
communiction_score=[]
location_score = []
check_in_score=[]
cleaning_score = []
value_for_money_score = []
Tv=[]
wifi = []
elevator=[]
privet_parking=[]
free_street_parking=[]
Refrigerator=[]
air_conditioning=[]
kitchen=[]
is_pets_allowed = []
workspace=[]
cameras=[]
```

# crawling

```python

for i in range(num_of_pages):
    main_page = driver.current_url
    ApartmentsUrl = CountPageUrls(driver)
    
    print("Moving page ..")
    print(f"Page number : {i+1}")
    
    for j in range(len(ApartmentsUrl)):
        print(f"apartment number {j+1} in this page")
        driver.get(ApartmentsUrl[j])
        time.sleep(3)
        
    
# Parsing the page to BeautifulSoup.
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")

    
# get Apartment's name
        try:
            names = soup.find('h1', {"class": "_fecoyn4"})
            apartments_names.append(names.get_text())
            print(f"name : {names.get_text()}")
        except:
            apartments_names.append(f"apartment in page {i} and number {j+1}")
            print(f"apartment in page {i} and number {j+1}")
            
        
# all rating components:
        rating_comp = soup.findAll('span', {"class": "_4oybiu"})
        if(len(rating_comp)!=0):
            cleaning_score.append(rating_comp[0].get_text())
            print(f"cleaning : {rating_comp[0].get_text()}")
            accuracy_score.append(rating_comp[1].get_text())
            print(f"accuracy : {rating_comp[1].get_text()}")
            communiction_score.append(rating_comp[2].get_text())
            print(f"communiction : {rating_comp[2].get_text()}")
            location_score.append(rating_comp[3].get_text())
            print(f"location : {rating_comp[3].get_text()}")
            check_in_score.append(rating_comp[4].get_text())
            print(f"check_in : {rating_comp[4].get_text()}")
            value_for_money_score.append(rating_comp[5].get_text())
            print(f"value_for_money : {rating_comp[5].get_text()}")
        else:
            cleaning_score.append(np.nan)
            accuracy_score.append(np.nan)
            communiction_score.append(np.nan)
            location_score.append(np.nan)
            check_in_score.append(np.nan)
            value_for_money_score.append(np.nan)
            
            
# get Number of Apartment's rating
        num_stars = soup.find('span', {"class": "_17p6nbba"})
        if(num_stars):
            overall_rating.append(num_stars.get_text()[0:-1])
            print(f"Stars : {num_stars.get_text()[0:-1]}")
        else:
            overall_rating.append(np.nan)
            print("Stars : nan")
            
         
 # get Apartment's price
 #split from stack overflow
        try:
            prices= soup.find('span',{"class":"_1k4xcdh"})
            prices_per_month.append(prices.text.split()[0][1:])
            print(f"Price : {prices.text.split()[0][1:]}")
        except:
            prices_per_month.append(np.nan)
            print(f"Price : nan ")
    
# get Apartment's number of guests
        try:
            guest = soup.find('div', {'class': "_tqmy57"})
            span2 = guest.findAll('span')
            guests = span2[0].text.split()[0]
            num_of_guests.append(guests)
            print(f"Guests : {guests}")
        except:
            num_of_guests.append(np.nan)
            print(f"Guests : nan")
            
            
            
# get Apartment's number of reviewers
        ratingSumStr= soup.findAll('button',{"class":"_11eqlma4"}) 
        if (len(ratingSumStr)>1):
            num_of_reviews.append(ratingSumStr[0].text.split()[0])
            print(f"Reviews : {ratingSumStr[0].text.split()[0]}")
        else:
            num_of_reviews.append("0")
            print("Reviews : 0") 
        
     
 # get if there is wifi in  the Apartment's
 # find by text from stack overflow

        try:
            if(driver.find_element(By.XPATH, '//div[text()="Wifi"]/following-sibling::div')):
                wifi.append("1")
                print("Wifi : 1")
        except:
            wifi.append("0")
            print("Wifi : 0")
            
            
# get if there is Air conditioning in the Apartment's
        try:
            if(driver.find_element(By.XPATH, '//div[text()="Air conditioning"]/following-sibling::div')):
                air_conditioning.append("1")
                print("Air conditioning : 1")
        except:
            air_conditioning.append("0")
            print("Air conditioning : 0")
           
        
# get if there is Kitchen in the Apartment's
        try:
            if(driver.find_element(By.XPATH, '//div[text()="Kitchen"]/following-sibling::div')):
                kitchen.append("1")
                print("Kitchen : 1")
        except:
            kitchen.append("0")
            print("Kitchen : 0")
            
            
# get if Pets allowed in the Apartment's
        try:
            if(driver.find_element(By.XPATH, '//div[text()="Pets allowed"]/following-sibling::div')):
                is_pets_allowed.append("1")
                print("Pets allowed : 1")
        except:
            is_pets_allowed.append("0")
            print("Pets allowed : 0")
            
            
# get if there is Free parking near the Apartment's      
        try:
            if(driver.find_element(By.XPATH, '//div[text()="Free parking on premises"]/following-sibling::div')):
                privet_parking.append("1")
                print("Privet parking : 1")
            
        except:
            privet_parking.append("0")
            print("Privet parking : 0")
        try:
            if(driver.find_element(By.XPATH, '//div[text()="Free street parking"]/following-sibling::div')):
                free_street_parking.append("1")
                print("Free street parking : 1")
        except:
            free_street_parking.append("0")
            print("Free street parking : 0")
            
            
# get if there is TV in the Apartment's            
        try:
            if(driver.find_element(By.XPATH, '//div[text()="TV"]/following-sibling::div')):
                Tv.append("1")
                print("TV : 1")
        except:
            Tv.append("0")
            print("TV : 0")


# get if there is Elavator in the Apartment's            
        try:
            if(driver.find_element(By.XPATH, '//div[text()="Elavator"]/following-sibling::div')):
                elevator.append("1")
                print("Elavator : 1")
        except:
            elevator.append("0")
            print("Elavator : 0")       
     
    
# get if there is Refrigerator in the Apartment's            
        try:
            if(driver.find_element(By.XPATH, '//div[text()="Refrigerator"]/following-sibling::div')):
                Refrigerator.append("1")
                print("Refrigerator : 1")
        except:
            Refrigerator.append("0")
            print("Refrigerator : 0") 
            
            
# get if there is Dedicated workspace in the Apartment's            
        try:
            if(driver.find_element(By.XPATH, '//div[text()="Dedicated workspace"]/following-sibling::div')):
                workspace.append("1")
                print("Dedicated workspace : 1")
        except:
            workspace.append("0")
            print("Dedicated workspace : 0")
            
            
# get if there is Security cameras on property's            
        try:
            if(driver.find_element(By.XPATH, '//div[text()="Security cameras on property"]/following-sibling::div')):
                cameras.append("1")
                print("Security cameras on property : 1")
        except:
            cameras.append("0")
            print("Security cameras on property : 0")
            
            
    
        print("*******************************************")
#When the selenuom will arrive to the bottom of the page 
#the selenum will go back to the main page and go to next page 
    driver.get(main_page)
    time.sleep(4)
#click on next page button from Github 
    if i != num_of_pages - 1  :
        next_page=driver.find_element(By.CLASS_NAME, "_1bfat5l")
        print(next_page)
        next_page.click()
        time.sleep(3)
#The end of the crawling
driver.quit()
```
